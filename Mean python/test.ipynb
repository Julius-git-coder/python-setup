```python
import numpy as np

import matplotlib.pyplot as plt

# Data representing the earnings of market women (in dollars)
earnings = [50, 60, 55, 70, 65, 80, 75, 90, 85, 100]

# Calculate mean and median
mean_earnings = np.mean(earnings)
median_earnings = np.median(earnings)

# Display the data
plt.figure(figsize=(10, 6))
plt.bar(range(len(earnings)), earnings, color='skyblue', label='Earnings')
plt.axhline(mean_earnings, color='green', linestyle='--', label=f'Mean: {mean_earnings:.2f}')
plt.axhline(median_earnings, color='orange', linestyle='--', label=f'Median: {median_earnings:.2f}')
plt.title('Earnings of Market Women')
plt.xlabel('Market Women')
plt.ylabel('Earnings ($)')
plt.legend()
plt.show()
```
