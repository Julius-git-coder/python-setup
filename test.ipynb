```python
import random
import string

def generate_random_users(num_users=10, username_length=8):
    users = []

    for _ in range(num_users):
        username = ''.join(
            random.choices(string.ascii_letters + string.digits, k=username_length)
        )
        users.append(username)

    return users


# Example usage
random_users = generate_random_users(10)
print(random_users)
```
